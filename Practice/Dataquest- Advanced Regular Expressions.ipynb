```python
import pandas as pd
import re
```

```python
hn = pd.read_csv('hacker_news.csv')
titles = hn['title']
pattern = r'SQL'
sql_counts = titles.str.contains(pattern, flags=re.I).value_counts()
sql_counts
```

```python
hn_sql = hn[hn['title'].str.contains(r'\w+SQL',flags=re.I)].copy()
hn_sql
```

```python
hn_sql['flavor'] = hn_sql['title'].str.extract(r'(\w+SQL)',flags=re.I,expand=False)
hn_sql['flavor'].head()
```

```python
hn_sql['flavor'] =hn_sql['flavor'].str.lower()
sql_pivot = hn_sql.pivot_table(index='flavor',values='num_comments',aggfunc='mean')
sql_pivot
```

```python
pattern = r'[Pp]ython ([\d\.]+)'
py_versions = titles.str.extract(pattern,expand=False)
py_versions_freq = dict(py_versions.value_counts())
print(py_versions.value_counts())
print(py_versions_freq)
```

```python
def first_10_matches(pattern):
    """
    Return the first 10 story titles that match
    the provided regular expression
    """
    all_matches = titles[titles.str.contains(pattern)]
    first_10 = all_matches.head(10)
    return first_10

pattern = r'\b[Cc]\b[^+.]'
first_10 = first_10_matches(pattern)
first_10
```

```python
pattern = r'(?<!Series\s)\b[Cc]\b(?![+.]|\.$)'
#c_extract = titles.str.extract(pattern,expand=False) #ValueError: pattern contains no capture groups
#print(c_extract)
c_pattern =  titles.str.contains(pattern).value_counts()
print(c_pattern)
c_mentions = titles.str.contains(pattern).sum()
print(c_mentions)
```

```python
pattern = r'\b(\w+)\s\1\b'
repeated_words = titles[titles.str.contains(pattern)]
repeated_words
```

```python
email_variations = pd.Series(['email', 'Email', 'e Mail',
                        'e mail', 'E-mail', 'e-mail',
                        'eMail', 'E-Mail', 'EMAIL'])

pattern = r"\be[\s-]*mail"

email_uniform = email_variations.str.replace(pattern, "email", flags=re.I)
titles_clean = titles.str.replace(pattern, "email",flags=re.I)
titles_clean
```
