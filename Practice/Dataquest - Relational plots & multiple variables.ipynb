```python
import pandas as pd
import seaborn as sns

housing = pd.read_csv('housing.csv')
housing.head()
```

```python
housing.info()
```

```python
import matplotlib.pyplot as plt
sns.set_theme() #visually graph use matplotlib default, this allow to switch to seaborn defaults
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice')
plt.show()
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual', palette='Spectral')
plt.show()
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual', palette='Spectral', size='Garage Area', sizes=(1,500))
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual', palette='Spectral', size='Garage Area', sizes=(1,300), style='Rooms',markers=['^','o'])
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual', palette='Spectral', size='Garage Area', sizes=(1,300), style='Rooms',markers=['^','o'], col='Year')
```

In this lesson, we learned the following:

Building relational plots with Seaborn.  
Visually representing multiple variables using x- and y-coordinates, color, size, shape, and spatial separation.  
Identifying patterns by examining a relational plot.
