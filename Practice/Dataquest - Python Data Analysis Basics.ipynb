```python
from csv import reader

opened_file = open('artworks_clean.csv', encoding='UTF-8')
read_file = reader(opened_file)
moma = list(read_file)
moma = moma[1:]
```

```python
for row in moma:
    birth_date = row[3]
    if birth_date != "":
        birth_date = int(birth_date)
    row[3] = birth_date
```

```python
for row in moma:
    death_date = row[4]
    if death_date != "":
        death_date = int(death_date)
    row[4] = death_date
```

```python
for row in moma:
    date = row[6]
    if date != "":
        date = int(date)
    row[6] = date
```

```python
ages = []
for row in moma:
    date = row[6]
    birth = row[3]
    if type(birth) == int:
        age = birth
    else:
        age = 0
    ages.append(age)
```

```python
final_ages = []
for age in ages:
    if age > 20:
        final_age = age
    else:
        final_age = "Unknown"
    final_ages.append(final_age)
```

```python
decades = []

for age in final_ages:
    if age == "Unknown":
        decade = age
    else:
        decade = str(age)
        decade = decade[:-1]+"0s"
    decades.append(decade)
```

```python
decade_frequency = {}

for x in decades:
    decade_frequency[x] = decade_frequency.get(x,0)+1

print(decade_frequency)
```
