```python
import pandas as pd
import matplotlib.pyplot as plt

bike = pd.read_csv('day.csv')
plt.scatter(bike['workingday'], bike['casual'])
plt.title('Day vs Casual')
plt.xlabel('Work day')
plt.ylabel('Casual')
plt.show()
```

```python
plt.scatter(bike['workingday'], bike['registered'])
plt.title('Day vs Registered')
plt.show()
```

```python
working_days = ['Non-Working Day', 'Working Day']
registered_avg = [2959, 3978]
plt.bar( working_days, registered_avg)
plt.show()
```

```python
weekday_averages = bike.groupby('weekday').mean()[['casual','registered']].reset_index()
```

```python
plt.bar(weekday_averages['weekday'], weekday_averages['registered'])
plt.xticks(ticks=[0,1,2,3,4,5,6], labels=['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'], rotation=30)
plt.show()
```

```python
unique_values = [1, 2, 3, 4]
weather_2011 = [226, 124, 15, 0]
weather_2012 = [237, 123, 6, 0]

plt.bar(unique_values, weather_2012)
plt.xticks(ticks=unique_values)
plt.title('Weather Patterns: 2012')
plt.ylabel('Frequency')
plt.xlabel('Unique Values')
plt.show()
```

```python
bike['registered'].value_counts(bins=10).sort_index()
```

```python
bike['casual'].value_counts(bins=10).sort_index()
```

```python
plt.hist(bike['casual'])
plt.show()
```
