```python
import pandas as pd

bike = pd.read_csv('day.csv')
bike.head()
```

```python
bike.info()
```

```python
import matplotlib.pyplot as plt

bike['dteday'] = pd.to_datetime(bike['dteday'])
plt.plot(bike['dteday'],bike['temp'])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(bike['windspeed'],bike['cnt'])
plt.xlabel('Wind Speed')
plt.ylabel('Bikes Rented')
plt.title('Windspeed vs Bike rentals')
plt.show()
```

```python
plt.scatter(bike['atemp'],bike['registered'])
plt.xticks(rotation=45)
plt.show()
```

```python
bike['atemp'].corr(bike['registered'])
```

```python
bike['atemp'].corr(bike['temp'])
```

```python
plt.scatter(bike['temp'],bike['atemp'])
plt.xlabel('Air Temp')
plt.ylabel('Feeling temp')
plt.show()
```
