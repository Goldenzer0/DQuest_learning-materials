```python
import numpy as np
import pandas as pd

laptops = pd.read_csv('laptops.csv',encoding='Latin-1')
laptops.info()
```

```python
new_columns = []
#remove whitespaces in columns names
for column in laptops.columns:
    column = column.strip()
    new_columns.append(column)
laptops.columns = new_columns
```

```python
laptops.info()
```

```python
def clean_col(col):
    col = col.strip()
    col = col.replace('Operating System','os')
    col = col.replace(' ','_')
    col = col.replace('(','')
    col = col.replace(')','')
    col = col.lower()
    return col
new_columns = []
for col in laptops.columns:
    col = clean_col(col)
    new_columns.append(col)

laptops.columns = new_columns
```

```python
laptops['ram'] = laptops['ram'].str.replace('GB','')
unique_ram = laptops['ram'].unique()
unique_ram
```

```python
laptops['ram']=laptops['ram'].astype(int)
laptops.dtypes
```

```python
laptops.rename({'ram':'ram_gb'},axis=1,inplace=True)
```

```python
laptops['ram_gb'].describe()
```

```python
laptops.head()
```

```python
laptops['gpu_manufacturer'] = laptops['gpu'].str.split().str[0]
laptops['cpu_manufacturer'] = laptops['cpu'].str.split().str[0]
laptops['cpu_manufacturer'].value_counts()
```

```python
laptops.head()
```

```python
mapping_dict = {
    'Android': 'Android',
    'Chrome OS': 'Chrome OS',
    'Linux': 'Linux',
    'Mac OS': 'macOS',
    'No OS': 'No OS',
    'Windows': 'Windows',
    'macOS': 'macOS'
}
laptops['os'] = laptops['os'].map(mapping_dict)
```

```python
laptops_no_null_rows = laptops.dropna(axis=0)
laptops_no_null_cols = laptops.dropna(axis=1)
```

```python
value_counts_before = laptops.loc[laptops["os_version"].isnull(), "os"].value_counts()
laptops.loc[laptops["os"] == "macOS", "os_version"] = "X"
laptops.loc[laptops['os'] == 'No OS', 'os_version'] = 'Version Unknown'
value_counts_after = laptops.loc[laptops['os_version'].isnull(),'os'].value_counts()
print('Before:\n',value_counts_before)
print('After:\n',value_counts_after)
```

```python
laptops["weight"] = laptops["weight"].str.replace("kgs","")
laptops['weight'] = laptops['weight'].str.replace("kg","")
laptops['weight'] = laptops['weight'].astype(float)
laptops.rename({"weight": "weight_kg"}, axis=1, inplace=True)
laptops.to_csv('laptops_cleaned.csv',index=False)
```

-Convert the price_euros column to a numeric dtype.  
-Extract the screen resolution from the screen column.  
-Extract the processor speed from the cpu column.  

```python
laptops['price_euros'] = laptops['price_euros'].str.replace(',','')
laptops['price_euros'] = laptops['price_euros'].astype(int)
laptops.dtypes
```

```python
laptops['screen_resolution'] = laptops['screen'].str[-9:]
laptops.head()
```

```python
laptops['cpu_speed'] = laptops['cpu'].str[-6:]
laptops.head()
```

Are laptops made by Apple more expensive than those made by other manufacturers?

```python
top_price_by_manufacturer ={}
unique_manufacturer =  laptops['manufacturer'].unique()
for company in unique_manufacturer:
    is_company = laptops['manufacturer']==company
    companies = laptops.loc[is_company]
    top_price = companies.sort_values('price_euros',ascending=False).iloc[0]['price_euros']
    top_price_by_manufacturer[company] = top_price
top_price_by_manufacturer
```

```python
from heapq import nlargest
top_5_price = nlargest(5,top_price_by_manufacturer,key=top_price_by_manufacturer.get)
top_5_price
```

```python
sorted(top_price_by_manufacturer, key=top_price_by_manufacturer.get, reverse=True)[:5]
```
